#Faculdades Publicas <hr />

> Importação dos arquivos

```python
#Copiar caminho do drive ...
arquivoFacPublicas = pandas.read_csv("/content/drive/MyDrive/Projeto Teste/Projeto Final/DADOS/Cópia de faculdades_publicas.csv")
df_faculdades_publicas = pandas.DataFrame(arquivoFacPublicas)
```

> Modelagem dados

```python
media_total = df_faculdades_publicas['brancos '].mean() + df_faculdades_publicas['negros '].mean() + df_faculdades_publicas['pardos'].mean()
total_alunos_eng = df_faculdades_publicas['brancos '].max() + df_faculdades_publicas['negros '].max() + df_faculdades_publicas['pardos'].max()
porcentagem_brancos = (df_faculdades_publicas['brancos '].mean() / media_total) *100
porcentagem_negros = (df_faculdades_publicas['negros '].mean() / media_total) *100
porcentagem_pardos = (df_faculdades_publicas['pardos'].mean() / media_total) *100
```

> Exibição dos dados

```python
df_faculdades_publicas.plot(kind='bar', title='Numero de matriculas em faculdades publicas por cor ou raça\n',figsize=(12,5))
```

```python
print('\n----------- Apresentação dos dados ----------')
print(df_faculdades_publicas)


print('\n----------- Maximo valor ----------')
print(df_faculdades_publicas.max())

print('\n----------- Relação entre valores ----------')
print(f'=> Total de pessoas, de qualquer raça, matriculadas no curso de engenharia e construção : "{round(total_alunos_eng)}"')
print(f'=> A media de pessoas, de qualquer raça, ingressadas na faculdade, em qualquer curso, é: "{round(media_total)}" pessoas')
print(f'=> A porcentagem de brancos é de: {round(porcentagem_brancos,2)}%')
print(f'=> A porcentagem de negros é de: {round(porcentagem_negros,2)}%')
print(f'=> A porcentagem de pardos é de: {round(porcentagem_pardos,2)}%\n\n')

racas = ["Brancos", "Negros", "Pardos"]
porcent = [porcentagem_brancos,porcentagem_negros, porcentagem_pardos]

plt.pie(porcent, labels=racas, autopct='%1.2f%%')
plt.title(f'Taxa de alunos matriculados \nem faculdades públicas')

print('----------- Representação em grafico ----------')
plt.show()
```

##Conclusão previa:
`Neste cenario, é feito uma analise sobre os alunos em faculdades publicas, de tal maneira a explorar possiveis discrepancias em relação a dificuldade de acesso a uma instituição privada por parte do negro em relação ao branco.`

`Com os dados, foram evidenciados a quantidade de alunos matriculados por cor e raça. Foram listados 8 cursos diversos para realizar o estudo`

`Foi feito uma media sobre a quantidade de alunos matriculados, em qualquer curso, levando em consideração cor e raça. O resultado obtido foi uma media de: 55.235 alunos brancos matriculados, 7.653 alunos negros matriculados e 34.493 alunos pardos matriculados;`

####Isto resulta em uma media de:
  >86448 pessoas matriculadas por curso

####Sendo desta quantidade,a taxa de alunos matriculas por cor ou raça
  >Alunos brancos matriculados de 54,86%
  
  >Alunos negros matriculados de 10,22% 

  >Alunos pardos matriculados de 34.92%

`Ou seja, apenas aproximadamente 10% dos alunos matriculados são negros, e mesmo com a miscigenação racial, não chega a equiparar com a quantidade de alunos brancos com diretos de acesso a uma faculdade privada`

#Ocupações Formais <hr />


> Importação dos arquivos

```python
arquivo_ocupacoes_formais = pandas.read_csv('/content/drive/MyDrive/Projeto Teste/Projeto Final/DADOS/Cópia de ocupacoesfi_renovado.csv', sep=",")
df_ocupacoes_formais = pandas.DataFrame(arquivo_ocupacoes_formais)
```

> Modelagem de dados:

```python
taxa_pessoas_negras = list((df_ocupacoes_formais['preta'] / df_ocupacoes_formais['total']) * 100)
taxa_pessoas_brancas = list((df_ocupacoes_formais['branca'] / df_ocupacoes_formais['total']) * 100)
taxa_pessoas_pardas = list((df_ocupacoes_formais['parda'] / df_ocupacoes_formais['total']) * 100)
```

> Exibição dos dados

```python
arquivo_ocupacoes_formais.plot(kind='bar',title='Taxa de pessoas com ocupações formais\n por cor ou raça\n', figsize=(4,3)) 
```

```python
print('\n----------- Apresentação dos dados ----------')
print(df_ocupacoes_formais)

print('\n----------- Maximo valor ----------')
print(df_ocupacoes_formais.max())

print('\n----------- Relação entre valores ----------')
print(f'Taxa de pessoas negras em ocupações formais: {round(taxa_pessoas_negras[0],2)}%')
print(f'Taxa de pessoas brancas em ocupações formais: {round(taxa_pessoas_brancas[0],2)}%')
print(f'Taxa de pessoas pardas em ocupações formais: {round(taxa_pessoas_pardas[0],2)}%\n\n')

r_ocup_formais = ["Negras", "Brancas", "Pardos"]
porcent = [taxa_pessoas_negras[0], taxa_pessoas_brancas[0], taxa_pessoas_pardas[0]]

plt.pie(porcent, labels=r_ocup_formais, autopct='%1.2f%%')
plt.title('Ocupações Formais')

print('----------- Representação em grafico ----------')
plt.show()
```

>Percentual de pessoas negras em ocupações formais: 9.39%
