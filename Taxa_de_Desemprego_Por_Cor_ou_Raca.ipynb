# Homicidios <hr />

```python
# Importando CSV (OBS: Substituir caminho no arquivo principal)
dataset_homicidios = pandas.read_csv("/content/drive/MyDrive/datasets/csv/homicidios.csv")

df_homicidios = pandas.DataFrame(dataset_homicidios)

print(df_homicidios)
```

```python
taxa_homicidios_brancas = df_homicidios.iloc[0]['Taxa']
taxa_homicidios_negras = df_homicidios.iloc[1]['Taxa']
taxa_homicidios_pardas = df_homicidios.iloc[2]['Taxa']

total_homicidios = df_homicidios.iloc[0]['absoluto'] + df_homicidios.iloc[1]['absoluto'] + df_homicidios.iloc[2]['absoluto']
```

```python
print('\n----------- Descrição estatistica pandas ----------')
print(df_homicidios.head())


print('\n----------- Maximo valor ----------')
print(df_homicidios.max())

print('\n----------- Relação entre valores ----------')
print(f'=> Total de homicidios em um ano, independente da raça é de : {total_homicidios * 100000} pessoas')
print(f'=> A porcentagem de homicídios de brancos é de: {taxa_homicidios_brancas}%')
print(f'=> A porcentagem de homicídios de negros é de: {taxa_homicidios_negras}%')
print(f'=> A porcentagem de homicídios de de pardos é de: {taxa_homicidios_pardas}%')


racas = ["Brancos", "Negros", "Pardos"]
porcent = [taxa_homicidios_brancas, taxa_homicidios_negras, taxa_homicidios_pardas]

plt.pie(porcent, labels=racas, autopct='%1.2f%%')
plt.title('Homicidios')
print('\n----------- Representação em grafico ----------')
print()
plt.show()
```

## Conclusão previa:

`Neste cenario, é feito uma analise sobre os homicídios em um ano. ` 

` Com os dados, foram evidenciados que a taxa de homicídios de negros e pardos é bem mais alta que a de brancos. `

` O número absoluto de homicídios em 2020 foi de 4862500 pessoas. `

- Homicídios de brancos foram cerca de 11,50%  
- Homicídios de negros foram cerca de 21,90% 
- Homicídios de pardos foram cerca de 34,10% 

` Ou seja, mais de 50% dos homicídios são de pessoas negras ou pardas, o que reforça a teste. `


#Faculdades Publicas <hr />

> Importação dos arquivos

```python
#Copiar caminho do drive ...
arquivoFacPublicas = pandas.read_csv("/content/drive/MyDrive/datasets/BACKUP/DADOS/Cópia de faculdades_publicas.csv")
df_faculdades_publicas = pandas.DataFrame(arquivoFacPublicas)
```

```python
df_faculdades_publicas.plot(kind='bar', title='Numero de matriculas em faculdades publicas por cor ou raça\n',figsize=(12,5))
```

> Modelagem dados

```python
media_total = df_faculdades_publicas['brancos '].mean() + df_faculdades_publicas['negros '].mean() + df_faculdades_publicas['pardos'].mean()
total_alunos_eng = df_faculdades_publicas['brancos '].max() + df_faculdades_publicas['negros '].max() + df_faculdades_publicas['pardos'].max()
porcentagem_brancos = (df_faculdades_publicas['brancos '].mean() / media_total) *100
porcentagem_negros = (df_faculdades_publicas['negros '].mean() / media_total) *100
porcentagem_pardos = (df_faculdades_publicas['pardos'].mean() / media_total) *100
```

> Exibição dos dados

```python
print('\n----------- Descrição estatistica pandas ----------')
print(df_faculdades_publicas.head())


print('\n----------- Maximo valor ----------')
print(df_faculdades_publicas.max())

print('\n----------- Relação entre valores ----------')
print(f'=> Total de pessoas, de qualquer raça, matriculadas no curso de engenharia e construção : "{round(total_alunos_eng)}"')
print(f'=> A media de pessoas, de qualquer raça, ingressadas na faculdade, em qualquer curso, é: "{round(media_total)}" pessoas')
print(f'=> A porcentagem de brancos é de: {round(porcentagem_brancos,2)}%')
print(f'=> A porcentagem de negros é de: {round(porcentagem_negros,2)}%')
print(f'=> A porcentagem de pardos é de: {round(porcentagem_pardos,2)}%\n')

racas = ["Brancos", "Negros", "Pardos"]
porcent = [porcentagem_brancos,porcentagem_negros, porcentagem_pardos]

plt.pie(porcent, labels=racas, autopct='%1.2f%%')
plt.title(f'Taxa de alunos matriculados \n em faculdades públicas')

print('----------- Representação em grafico ----------\n')
plt.show()
```

##Conclusão previa:
`Neste cenario, é feito uma analise sobre os alunos em faculdades publicas, de tal maneira a explorar possiveis discrepancias em relação a dificuldade de acesso a uma instituição privada por parte do negro em relação ao branco.`

`Com os dados, foram evidenciados a quantidade de alunos matriculados por cor e raça. Foram listados 8 cursos diversos para realizar o estudo`

`Foi feito uma media sobre a quantidade de alunos matriculados, em qualquer curso, levando em consideração cor e raça. O resultado obtido foi uma media de: 55.235 alunos brancos matriculados, 7.653 alunos negros matriculados e 34.493 alunos pardos matriculados;`
