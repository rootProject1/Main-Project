# Habitantes no Brasil


> Importação dos Arquivos

```python
#Importando CSV (OBS: Substituir caminho no arquivo principal)
dataset_metropole_absoluto = pandas.read_csv('/content/habitantes_por_regiao.csv')
dataset_metropole_porcentagem = pandas.read_csv('/content/habitantes-geral.csv')
dataset_total = pandas.read_csv("/content/habitantes_total.csv")

#Transformando os arquivos .csv em Dataframes, os quais serão modelados nesta seção
df_total = pandas.DataFrame(dataset_total)
df_metropole_absoluto = pandas.DataFrame(dataset_metropole_absoluto)
df_metropole_porcentagem = pandas.DataFrame(dataset_metropole_porcentagem)
```

> Modelagem dados

```python
taxa_pessoas_norte = df_metropole_porcentagem.iloc[1]
taxa_pessoas_nordeste = df_metropole_porcentagem.iloc[2]
taxa_pessoas_sudeste = df_metropole_porcentagem.iloc[3]
taxa_pessoas_sul = df_metropole_porcentagem.iloc[4]
taxa_pessoas_centro_oeste = df_metropole_porcentagem.iloc[5]

taxa_brancas_norte = df_metropole_porcentagem.iloc[1]['brancas']
taxa_brancas_nordeste = df_metropole_porcentagem.iloc[2]['brancas']
taxa_brancas_sudeste = df_metropole_porcentagem.iloc[3]['brancas']
taxa_brancas_sul = df_metropole_porcentagem.iloc[4]['brancas']
taxa_brancas_centro_oeste = df_metropole_porcentagem.iloc[5]['brancas']

taxa_negras_norte = df_metropole_porcentagem.iloc[1]['negras']
taxa_negras_nordeste = df_metropole_porcentagem.iloc[2]['negras']
taxa_negras_sudeste = df_metropole_porcentagem.iloc[3]['negras']
taxa_negras_sul = df_metropole_porcentagem.iloc[4]['negras']
taxa_negras_centro_oeste = df_metropole_porcentagem.iloc[5]['negras']

taxa_pardas_norte = df_metropole_porcentagem.iloc[1]['pardas']
taxa_pardas_nordeste = df_metropole_porcentagem.iloc[2]['pardas']
taxa_pardas_sudeste = df_metropole_porcentagem.iloc[3]['pardas']
taxa_pardas_sul = df_metropole_porcentagem.iloc[4]['pardas']
taxa_pardas_centro_oeste = df_metropole_porcentagem.iloc[5]['pardas']

taxa_total_brancas = df_total.iloc[0]['Brancas']
taxa_total_negras = df_total.iloc[0]['Negras']
taxa_total_pardas = df_total.iloc[0]['Pardas']
```

> Exibição dos dados

```python
#Plotagem do grafico de miscigenação por região
df_metropole_absoluto.plot(kind='bar', title='Miscigeração no brasil distinguido por região\nx1000 habitantes',figsize=(12,4))
barWidth = 0.05
plt.xticks([r + barWidth for r in range(len(df_metropole_absoluto))], ['Norte', 'Nordeste', 'Sudeste', 'Sul', 'Centro-Oeste'])
print()
```

```python
print('\n----------- Apresentação pandas ----------')
print(df_metropole_porcentagem)

print('\n----------- Relação entre valores ----------')
print(f'=> A porcentagem de pessoas brancas no Brasil é de : {taxa_total_brancas}%')
print(f'=> A porcentagem de pessoas negras no Brasil é de : {taxa_total_negras}%')
print(f'=> A porcentagem de pessoas pardas no Brasil é de : {taxa_total_pardas}%')

print('\n----------- Nordeste ----------')
print(f'=> A porcentagem de pessoas brancas no Nordeste é de: {taxa_brancas_nordeste}%')
print(f'=> A porcentagem de pessoas negras no Nordeste é de: {taxa_negras_nordeste}%')
print(f'=> A porcentagem de pessoas pardas no Nordeste é de: {taxa_pardas_nordeste}%')

print('\n----------- Norte ----------')
print(f'=> A porcentagem de pessoas brancas no Norte é de: {taxa_brancas_norte}%')
print(f'=> A porcentagem de pessoas negras no Norte é de: {taxa_negras_norte}%')
print(f'=> A porcentagem de pessoas pardas no Norte é de: {taxa_pardas_norte}%')

print('\n----------- Centro-Oeste ----------')
print(f'=> A porcentagem de pessoas brancas no Centro-Oeste é de: {taxa_brancas_centro_oeste}%')
print(f'=> A porcentagem de pessoas negras no Centro-Oeste é de: {taxa_negras_centro_oeste}%')
print(f'=> A porcentagem de pessoas pardas no Centro-Oeste é de: {taxa_pardas_centro_oeste}%')

print('\n----------- Sudeste ----------')
print(f'=> A porcentagem de pessoas brancas no Sudeste é de: {taxa_brancas_sudeste}%')
print(f'=> A porcentagem de pessoas negras no Sudeste é de: {taxa_negras_sudeste}%')
print(f'=> A porcentagem de pessoas pardas no Sudeste é de: {taxa_pardas_sudeste}%')

print('\n----------- Sul ----------')
print(f'=> A porcentagem de pessoas brancas no Sul é de: {taxa_brancas_sul}%')
print(f'=> A porcentagem de pessoas negras no Sul é de: {taxa_negras_sul}%')
print(f'=> A porcentagem de pessoas pardas no Sul é de: {taxa_pardas_sul}%')

print('\n----------- Representação em grafico ----------')
racas = ["Brancos", "Negros", "Pardos"]
porcent = [taxa_total_brancas,taxa_total_negras, taxa_total_pardas]

plt.pie(porcent, labels=racas, autopct='%1.2f%%')
plt.title(f'Taxa de miscigenação de raças no Brasil')
plt.show()
```

>A taxa de pessoas negras no Brasil é cerca de 9,10%;

#Faculdades Publicas <hr />

> Importação dos arquivos

```python
# Importando CSV (OBS: Substituir caminho no arquivo principal)
arquivoFacPublicas = pandas.read_csv("/content/matriculados_faculdades_publicas.csv")

#Transformando o arquivo .csv em Dataframe, o qual será modelado nesta seção
df_faculdades_publicas = pandas.DataFrame(arquivoFacPublicas)
```

> Modelagem dados

```python
media_total = df_faculdades_publicas['brancos '].mean() + df_faculdades_publicas['negros '].mean() + df_faculdades_publicas['pardos'].mean()
total_alunos_eng = df_faculdades_publicas['brancos '].max() + df_faculdades_publicas['negros '].max() + df_faculdades_publicas['pardos'].max()
porcentagem_brancos = (df_faculdades_publicas['brancos '].mean() / media_total) *100
porcentagem_negros = (df_faculdades_publicas['negros '].mean() / media_total) *100
porcentagem_pardos = (df_faculdades_publicas['pardos'].mean() / media_total) *100
```

> Exibição dos dados

```python
#Plotagem do grafico de numero de matriculas em faculdades publicas por cor ou raça
df_faculdades_publicas.plot(kind='bar', title='Numero de matriculas em faculdades publicas por cor ou raça\n',figsize=(10,5))
```

```python
print('\n----------- Apresentação dos dados ----------')
print(df_faculdades_publicas)


print('\n----------- Maximo valor ----------')
print(df_faculdades_publicas.max())

print('\n----------- Relação entre valores ----------')
print(f'=> Total de pessoas, de qualquer raça, matriculadas no curso de engenharia e construção : "{round(total_alunos_eng)}"')
print(f'=> A media de pessoas, de qualquer raça, ingressadas na faculdade, em qualquer curso, é: "{round(media_total)}" pessoas')
print(f'=> A porcentagem de brancos é de: {round(porcentagem_brancos,2)}%')
print(f'=> A porcentagem de negros é de: {round(porcentagem_negros,2)}%')
print(f'=> A porcentagem de pardos é de: {round(porcentagem_pardos,2)}%\n\n')

racas = ["Brancos", "Negros", "Pardos"]
porcent = [porcentagem_brancos,porcentagem_negros, porcentagem_pardos]

plt.pie(porcent, labels=racas, autopct='%1.2f%%')
plt.title(f'Taxa de alunos matriculados \nem faculdades públicas')

print('----------- Representação em grafico ----------')
plt.show()
```

##Conclusão previa:
`Neste cenario, é feito uma analise sobre os alunos em faculdades publicas, de tal maneira a explorar possiveis discrepancias em relação a dificuldade de acesso a uma instituição privada por parte do negro em relação ao branco.`

`Com os dados, foram evidenciados a quantidade de alunos matriculados por cor e raça. Foram listados 8 cursos diversos para realizar o estudo`

`Foi feito uma media sobre a quantidade de alunos matriculados, em qualquer curso, levando em consideração cor e raça. O resultado obtido foi uma media de: 55.235 alunos brancos matriculados, 7.653 alunos negros matriculados e 34.493 alunos pardos matriculados;`

####Isto resulta em uma media de:
  >86448 pessoas matriculadas por curso

####Sendo desta quantidade,a taxa de alunos matriculas por cor ou raça
  >Alunos brancos matriculados de 54,86%

  >Alunos negros matriculados de 10,22%

  >Alunos pardos matriculados de 34.92%

`Ou seja, apenas aproximadamente 10% dos alunos matriculados são negros, e mesmo com a miscigenação racial, não chega a equiparar com a quantidade de alunos brancos com diretos de acesso a uma faculdade privada`

`Destes, o curso com maior adesão foi de direito, com 279.846 pessoas no total, sendo que mais da metade dos alunos são brancos e pouco mais de 10% são negros, reforçando ainda mais a tese.`
